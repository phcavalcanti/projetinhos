Como mostrado acima, o binário de um número inteiro pode ser simplesmente obtido com o método bin(x). Mas se quiser remover o prefixo 0b da sua saída, pode utilizar a função format e formatar a saída.

A função format(value, format_spec) tem dois parâmetros - value e format_spec. Retornará a saída formatada de acordo com a função format_spec. Abaixo estão alguns exemplos de tipos de formatação que podem ser utilizados dentro dos espaços reservados:

Tipo de formatação	Papel

=	Coloca o sinal na posição mais à esquerda

b	Converte o valor em binário equivalente

o	Converte o valor para o formato octal

x	Converte o valor para o formato Hex

d	Converte o valor dado em decimal

E	Formato científico, com um E em maiúsculas

X	Converte o valor para o formato Hex em maiúsculas

E há muitos mais tipos de formatação disponíveis. Como queremos converter int em binário, será utilizado o tipo de formatação b.

Abaixo está o exemplo do código.

temp = format(10, "b")

print(temp)

````python
n = int(input('Insira um numero : '))
print()
base = int(input('informe o número correspondente para a base desejada :\n 1 - binário \n 2 - octal \n 3 - hexadecimal\n'))

if base == 1:
  bi = format(n,'b')
  print('O número {} em base binária equivale a {}'.format(n, bi))

elif base == 2:
  oct = format(n,'o')
  print('O número {} em base octal equivale a {}'.format(n, oct))

elif base == 3:
  hex = format(n,'x')
  print('O número {} em base hexadecimal equivale a {}'.format(n, hex))

````

Exercício Python 084: Faça um programa que leia nome e peso de várias pessoas, guardando tudo em uma lista. No final, mostre:

A) Quantas pessoas foram cadastradas.

B) Uma listagem com as pessoas mais pesadas.

C) Uma listagem com as pessoas mais leves.


````python
#raciocinio
#listas : dados , pacientes, pesados, leves
#while com append,clear e continua. Len para saber quantas pessoas foram cadastradas
#calculo de media para separar as pessoas entre mais pesadas e mais leves
#if para montar listas

pacientes = []
dados = []
pesados = []
leves = []
continua = 's'
pesoMedia = somaPeso = 0

while continua == 's':
  dados.append(str(input('Nome: ')))
  dados.append(float(input('Peso [kg]: ' )))
  pacientes.append(dados[:])
  dados.clear()
  print()

  continua = str(input('Deseja cadastrar mais pacientes ?  s/n : \n '))
  print()
  if continua == 'n':
    print('Cadastro encerrado')
    print()
    break
#[n] = paciente, [0] = nome, [1] = peso


print(f'Há {len(pacientes)} pacientes cadastrados')
print()
#calculo media de peso
for i in range(0,len(pacientes)):#para percorrer é preciso usar o range
  somaPeso += pacientes[i][1]#pacientes[i][1] acessa o peso dos pacientes

pesoMedia =somaPeso/len(pacientes)

#criando listas de separacao
for p in range(0,len(pacientes)):
  if pacientes[p][1] >= pesoMedia:#se for mais pesado que a media
    pesados.append(pacientes[p])
  else:
    leves.append(pacientes[p])#se for mais leve que a media

print(f'Para um peso médio de {pesoMedia :.2f} kg ')
print()
print(f'Os mais pesados são :{pesados}')
print()
print(f'Os mais leves são :{leves}')
````

Exercício Python 091: Crie um programa onde 4 jogadores joguem um dado e tenham resultados aleatórios. Guarde esses resultados em um dicionário em Python. No final, coloque esse dicionário em ordem, sabendo que o vencedor tirou o maior número no dado.

````python
#RACIOCINIO
#gerar numeros aleatorios que serao armazenados no dicionario resultados
#posvencedor, vencedor para o maior numero
from random import randint
from operator import itemgetter

resultados = dict()
vencedor=posvencedor =0

#jogo de dados
for d in range(0,4):
  resultados[f'dado {d+1}'] = randint(1,6)
print('Dados lançados:')

print(resultados)
print('=-'*20)

#verificacao do vencedor
for k, v in resultados.items():
  if v > vencedor :
    vencedor = v
    posvencedor = k


#apresentando em ordem:
ranking = dict()
ranking = sorted(resultados.items(), key = itemgetter(1), reverse = True )#o itemgetter ' vai ordenar pelo valor, por isso a key foi 1, se fosse pelo dado seria 0, usa reverse para apresentar em ordem decrescente, em forma de rankig



print(f'Ordem crescente : {sorted(resultados.values())}')
for i, v in enumerate(ranking):
  print(f'{i+1} ° lugar : {v[0]} - {v[1]}')

print() 

print(f'O vencedor foi {posvencedor}  que tirou {vencedor}')
````

Exercício Python 098: Faça um programa que tenha uma função chamada contador(), que receba três parâmetros: início, fim e passo. Seu programa tem que realizar três contagens através da função criada:

a) de 1 até 10, de 1 em 1
b) de 10 até 0, de 2 em 2
c) uma contagem personalizada

````python
from random import randint
from time import sleep
from operator import itemgetter#oredena lista de dicionarios
jogadores = dict()

jogadores['jogador_1'] = randint(1,6)
jogadores['jogador_2'] = randint(1,6)
jogadores['jogador_3'] = randint(1,6)
jogadores['jogador_4'] = randint(1,6)

print(jogadores)
#para colocar em ordem precisa ser criado outro dicionario
ranking = dict()
for k, v in jogadores.items():
  print(f'{k} tirou {v} no dado')
  sleep(1)
ranking = sorted(jogadores.items(), key = itemgetter(1), reverse = True )#colocou em ordem
print('*'*30)
for i, v in enumerate(ranking):
  print(f'{i+1}º lugar: {v}')
````
